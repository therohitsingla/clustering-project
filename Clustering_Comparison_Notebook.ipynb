# **Conclussion**

In comparing the results of K-Means, Hierarchical, and K-Means Shift clustering on the seeds dataset, several key trends emerge, particularly when considering the Silhouette score, Calinski-Harabasz index, and Davies-Bouldin score, which measure cluster cohesion, separation, and compactness.


---


K-Means Clustering
K-Means generally provides better performance when using transformations (T), achieving the highest Silhouette score (0.5343) with 4 clusters and transformation applied. This suggests that transforming the data significantly improves the quality of clustering. However, using transformations and normalizations together (T+N or T+N+PCA) tends to reduce performance across all metrics. Additionally, the Davies-Bouldin score, which measures cluster compactness (lower is better), is minimized at 0.5415 when using transformations.


---


Hierarchical Clustering
Hierarchical clustering generally performs worse than K-Means in terms of Silhouette scores, especially without data preprocessing (0.4066 for 3 clusters). Even with transformation and normalization, Hierarchical clustering achieves slightly lower Silhouette scores (around 0.5129) compared to K-Means. However, the Calinski-Harabasz index shows a high value (630.6559) with 5 clusters and transformations, indicating good separation of clusters under this method. The Davies-Bouldin score is comparable to K-Means, with 0.5312 being the best result when transformations are used.


---


K-Means Shift Clustering
K-Means Shift clustering shows a unique trend, with high Silhouette scores (0.6812) when normalization is applied, even though the Calinski-Harabasz and Davies-Bouldin scores are relatively low. Without any data preprocessing, the Silhouette score (0.4401) is comparable to other methods, but it underperforms in terms of the Calinski-Harabasz index (only 74.5656). This indicates that K-Means Shift is effective in terms of internal cohesion (Silhouette) but struggles with separation (Calinski-Harabasz).


---


Conclusion:-

K-Means clustering emerges as the most robust method, particularly when data transformation is applied, achieving a good balance of cohesion and separation. Hierarchical clustering can be competitive, especially when transformations are applied, but its performance is slightly weaker overall. K-Means Shift clustering, while showing excellent cohesion with normalization, struggles with cluster separation, making it less ideal for this dataset.
